```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data_plot = df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
df_market_data_plot
```

```python

# image of plot
#hvplot.save(df_market_data_plot, "./Images/market_data_plot.png")
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
df_market_data_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_data_improved = pd.DataFrame(df_market_data_scaled, columns=df_market_data.columns)

# Copy the crypto names from the original data


df_market_data_improved["coin_id"] = df_market_data.index
# Set the coinid column as index
df_market_data_improved = df_market_data_improved.set_index("coin_id")

# Display sample data
df_market_data_improved.head(10)
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
k
```

```python
# Create an empty list to store the inertia values


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list

inertia = []
k = list(range(1, 11))
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_improved)
    inertia.append(k_model.inertia_)

inertia
```

```python
# Create a dictionary with the data to plot the Elbow curve
# Create a DataFrame with the data to plot the Elbow curve


elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 

The goal of clustering is to create meaningful clusters where data points within a cluster are close together and far from other clusters. The elbow method, which plots explained variation against the number of clusters, is used to determine the optimal number of clusters, with 4 clusters being the best choice based on inertia values.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(df_market_data_improved)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predictions = model.predict(df_market_data_improved)


# Print the resulting array of cluster values.
predictions
```

```python
# Create a copy of the DataFrame
df_marketdata_predictions = df_market_data_improved.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters

df_marketdata_predictions["cluster_segment"] = predictions
# Display sample data
df_marketdata_predictions
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.


df_marketdata_predictions_plot = df_marketdata_predictions.hvplot.scatter(
    x = "price_change_percentage_24h",
    y = "price_change_percentage_7d",
    title = "Cryptocurrencies (K-Mean Clusters=4)",
    by = "cluster_segment",
    hover_cols = 'coin_id'
)

df_marketdata_predictions_plot
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
marketdata_pca = pca.fit_transform(df_market_data_improved)
# View the first five rows of the DataFrame. 
marketdata_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca.explained_variance_ratio_
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 

The fractions of explained variance for subsets of variables can be computed, and all three components explain 90% of the total variance. Explained variance is useful in assessing how important each component is in reducing the dimensionality of data, and using a 3rd Principal Component is beneficial in cases where the overall sum of variance ratios is low.

```python
# Create a new DataFrame with the PCA data.

# Creating a DataFrame with the PCA data
df_market_data_pca = pd.DataFrame(
    marketdata_pca,
    columns=["PCA1", "PCA2", "PCA3"]
)





# Copy the crypto names from the original data
df_market_data_pca['coin_id'] = df_market_data.index

# Set the coinid column as index
df_market_data_pca= df_market_data_pca.set_index('coin_id')


# Display sample data
df_market_data_pca
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_pca = list(range(1,11))
k_pca
```

```python
# Create an empy list to store the inertia values
inertia_pca = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list


for i in k_pca:
    k_model_pca = KMeans(n_clusters=i, random_state=1)
    k_model_pca.fit(df_market_data_pca)
    inertia_pca.append(k_model_pca.inertia_)
    
inertia_pca
```

```python
# Create a dictionary with the data to plot the Elbow curve

# Create a DataFrame with the data to plot the Elbow curve


elbow_data_pca = {"k": k_pca, "inertia": inertia_pca}

df_elbow_pca = pd.DataFrame(elbow_data_pca)
df_elbow_pca.head(10)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.

df_elbow_pca_plot = df_elbow_pca.hvplot.line(
    x = "k",
    y = "inertia",
    title = "PCA - Elbow Curve",
    xticks = k)
df_elbow_pca_plot
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**
  
  The best value is K=4. 

* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** 
  
 The best value for the number of clusters k is 4 based on the Elbow curve, which is the same as the best k value found using the original data. 
 
However, the Inertia at that point varies, with a lower value of 50 for 3 PCA compared to 79 for the original data, indicating that the strength of 'k' has improved with 3 PCA. The elbow on k=4 is more pronounced graphically.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data

model_pca.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
Predict_pca = model_pca.predict(df_market_data_pca)
# Print the resulting array of cluster values.
Predict_pca
```

```python
# Create a copy of the DataFrame with the PCA data
df_predictions_pca = df_market_data_pca.copy()

# Add a new column to the DataFrame with the predicted clusters
df_predictions_pca["pca_cluster"] = Predict_pca
# Display sample data
df_predictions_pca
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_predictions_pca_plot = df_predictions_pca.hvplot.scatter(
    x = "PCA1",
    y = "PCA2",
    title = "Cryptocurrencies (K-Mean PCA Clusters=4)",
    by = "pca_cluster",
    hover_cols = 'coin_id'
)
df_predictions_pca_plot
```
